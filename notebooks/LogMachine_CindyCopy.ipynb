```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import data
data = pd.read_csv("data/ModelData_2.csv")
data.head()
```

```python
#
testdata = data[['Calories/s', 'Fat/s', 'Sat_Fat/s', 'Trans_Fat/s', 'Mono_Fat/s', 'Poly_Fat/s',
                'Carbs/s', 'Fiber/s', 'Sugars/s','Protein/s', 'Cholesterol/s', 'Sodium/s', 'CarbScore', 'CholoScore', 'TotalFatScore','SugarScore','HealthyGradScore','HealthyScore']].copy()
testdata.head()                
```

```python
# Assign X (data) and y (target)

X = testdata.drop(['CarbScore', 'CholoScore', 'TotalFatScore','SugarScore','HealthyGradScore','HealthyScore'], axis=1)
y = testdata['HealthyGradScore']
print(X.shape, y.shape)
```

```python
#split data into test and tain
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
